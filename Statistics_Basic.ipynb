Q1: What is statistics, and why is it important?

A: Statistics is the science of collecting, organizing, analyzing, and interpreting data to make informed decisions. It’s important because it helps in understanding trends, making predictions, and drawing conclusions based on data.

Q2: What are the two main types of statistics?

A: The two main types are:

Descriptive Statistics – summarizing and describing data.

Inferential Statistics – drawing conclusions and making predictions based on samples.

Q3: What are descriptive statistics?

A: Descriptive statistics involve methods to summarize data using measures such as mean, median, mode, range, and standard deviation.

Q4: What is inferential statistics?

A: Inferential statistics use data from a sample to make generalizations or predictions about a population, often using probability theory.

Q5: What is sampling in statistics?

A: Sampling is the process of selecting a subset (sample) from a larger group (population) to draw conclusions about the entire population.

Q6: What are the different types of sampling methods?

A: Common methods include:

Simple Random Sampling

Stratified Sampling

Systematic Sampling

Cluster Sampling

Convenience Sampling

Purposive or Judgmental Sampling

Q7: What is the difference between random and non-random sampling?

A:

Random Sampling gives every member of the population an equal chance of being selected.

Non-Random Sampling involves selection based on judgment, convenience, or other non-random criteria.

Q8: Define and give examples of qualitative and quantitative data.

A:

Qualitative data: Descriptive, non-numeric (e.g., colors, names, opinions).

Quantitative data: Numerical (e.g., height, age, number of students).

Q9: What are the different types of data in statistics?

A:

Qualitative (Categorical): Nominal, Ordinal

Quantitative (Numerical): Discrete, Continuous

Q10: Explain nominal, ordinal, interval, and ratio levels of measurement.

A:

Nominal: Categories with no order (e.g., gender, colors)

Ordinal: Categories with order (e.g., rankings)

Interval: Ordered, with equal spacing, but no true zero (e.g., temperature in °C)

Ratio: Like interval, but with a true zero (e.g., weight, age)

Q11: What is the measure of central tendency?

A: It refers to values that describe the center of a data set – the most typical or average value. The main measures are mean, median, and mode.

Q12: Define mean, median, and mode.

A:

Mean: The average (sum of values ÷ number of values)

Median: The middle value when data is ordered

Mode: The most frequent value

Q13: What is the significance of the measure of central tendency?

A: It helps summarize a large dataset with a single value that represents the center or typical case, aiding comparisons and decision-making.

Q14: What is variance, and how is it calculated?

A: Variance measures how far data values spread out from the mean. It’s calculated as the average of the squared differences from the mean.

Q15: What is standard deviation, and why is it important?

A: Standard deviation is the square root of the variance. It tells how much the values deviate from the mean and is widely used because it’s in the same units as the data.

Q16: Define and explain the term range in statistics.

A: The range is the difference between the maximum and minimum values in a dataset. It gives a basic sense of data spread.

Q17: What is the difference between variance and standard deviation?

A: Variance is the average squared deviation from the mean, while standard deviation is the square root of variance, making it more interpretable in original data units.

Q18: What is skewness in a dataset?

A: Skewness measures the asymmetry of a distribution. A perfectly symmetric distribution has a skewness of 0.

Q19: What does it mean if a dataset is positively or negatively skewed?

A:

Positively skewed: Tail is longer on the right; most data is on the left.

Negatively skewed: Tail is longer on the left; most data is on the right.

Q20: Define and explain kurtosis.
A: Kurtosis measures the "tailedness" or peak of a distribution. High kurtosis indicates heavy tails and a sharp peak; low kurtosis means light tails and a flatter peak.

Q21: What is the purpose of covariance?

A: Covariance measures how two variables change together. A positive value means they increase together; a negative value means one increases while the other decreases.

Q22: What does correlation measure in statistics?

A: Correlation quantifies the strength and direction of the linear relationship between two variables, ranging from -1 to +1.

Q23: What is the difference between covariance and correlation?

A: Covariance measures joint variability but is affected by units. Correlation is a standardized version of covariance that is unitless and easier to interpret.

Q24: What are some real-world applications of statistics?

A: Statistics are used in:

Business (market research, quality control)

Medicine (drug testing, public health)

Government (census, policy-making)

Sports (performance analysis)

Education (test scoring, progress tracking)

```python
#1: How do you calculate the mean, median, and mode of a dataset?
import statistics as stats
import numpy as np
from scipy import stats as scipystats

data = [10, 20, 20, 30, 40, 50]

mean = np.mean(data)
median = np.median(data)
mode = scipystats.mode(data, keepdims=False)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode.mode)
```

```python
#2: Write a Python program to compute the variance and standard deviation of a dataset
import numpy as np

data = [15, 21, 23, 25, 30, 35]

variance = np.var(data, ddof=1)  # sample variance
std_dev = np.std(data, ddof=1)   # sample std deviation

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
#4: Implement sampling techniques like random sampling and stratified sampling
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.DataFrame({
    'ID': range(1, 11),
    'Group': ['A', 'A', 'B', 'B', 'A', 'B', 'B', 'A', 'A', 'B']
})

# Simple random sample
random_sample = df.sample(n=4)

# Stratified sample based on 'Group'
stratified_sample = df.groupby('Group', group_keys=False).apply(lambda x: x.sample(frac=0.5))

print("Random Sample:\n", random_sample)
print("\nStratified Sample:\n", stratified_sample)
```

```python
#5: Write a Python function to calculate the range of a dataset
def calculate_range(data):
    return max(data) - min(data)

sample_data = [4, 8, 15, 16, 23, 42]
print("Range:", calculate_range(sample_data))
```

```python
#12: Implement and compare simple random sampling and systematic sampling
import numpy as np

data = np.arange(1, 21)

# Simple Random Sampling
random_sample = np.random.choice(data, size=5, replace=False)

# Systematic Sampling
k = len(data) // 5
start = np.random.randint(0, k)
systematic_sample = data[start::k][:5]

print("Random Sample:", random_sample)
print("Systematic Sample:", systematic_sample)
```

```python
#13: Calculate the mean, median, and mode of grouped data
import pandas as pd
from scipy import stats

# Frequency distribution
grouped_data = {
    'class_interval': ['0-10', '10-20', '20-30', '30-40'],
    'frequency': [5, 8, 15, 7]
}

df = pd.DataFrame(grouped_data)
df['midpoint'] = df['class_interval'].apply(lambda x: (int(x.split('-')[0]) + int(x.split('-')[1])) / 2)
df['fx'] = df['midpoint'] * df['frequency']

mean = df['fx'].sum() / df['frequency'].sum()

print("Grouped Mean:", mean)
```

```python
#14: Simulate data using Python and calculate its central tendency and dispersion
data = np.random.normal(loc=50, scale=10, size=1000)

mean = np.mean(data)
median = np.median(data)
mode = scipystats.mode(data, keepdims=False).mode
std = np.std(data)
var = np.var(data)
rng = np.ptp(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", std)
print("Variance:", var)
print("Range:", rng)
```
